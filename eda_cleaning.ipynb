## 2. Data Cleaning

```python
df = df_raw.copy()

# Drop rank column
df = df.drop(columns=['Rk'], errors='ignore')

# Cast numeric columns
numeric_cols = ['Age','G','GS','MP','FG','FGA','FG%','3P','3PA','3P%',
                '2P','2PA','2P%','eFG%','FT','FTA','FT%',
                'ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS','Trp-Dbl']
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

print('Dtypes after cast:')
print(df.dtypes)
```

```python
# Players traded mid-season appear multiple times (once per team + a 'TOT' row).
# Keep only the TOT row so each player has one row per season.
def dedup_traded(group):
    if 'TOT' in group['Team'].values:
        return group[group['Team'] == 'TOT']
    return group

df = df.groupby(['Year', 'Player'], group_keys=False).apply(dedup_traded).reset_index(drop=True)

print(f'Shape after dedup: {df.shape}')
```

## 3. Fuzzy Join — NBA Totals + Draft Data

```python
# Estimate draft year for each totals player:
# The season Year is the end-year (e.g. 2004 = 2003-04 season),
# so a player drafted in summer 2003 first appears in Year 2004.
est_draft = (
    df.groupby('Player')['Year']
    .min()
    .reset_index()
    .rename(columns={'Year': 'first_season_year'})
)
est_draft['est_draft_year'] = est_draft['first_season_year'] - 1

print(f"Players with est_draft_year: {len(est_draft)}")
est_draft.head(10)
```

```python
import re, unicodedata
import jellyfish
from collections import defaultdict

def normalize_name(s):
    """Lowercase, strip accents, remove generational suffixes, strip punctuation."""
    s = unicodedata.normalize('NFKD', s).encode('ascii', 'ignore').decode('ascii').lower()
    s = re.sub(r'\b(jr\.?|sr\.?|ii|iii|iv)\b', '', s)
    s = re.sub(r'[^a-z\s]', '', s)
    return re.sub(r'\s+', ' ', s).strip()

def split_name(n):
    parts = n.split()
    if len(parts) >= 2:
        return parts[0], ' '.join(parts[1:])
    return n, n

def name_score(a, b):
    """Combined score weighting last name (0.65) more than first name (0.35).
    Scoring first/last separately prevents similar first names from masking
    completely different last names (e.g. Marcus Eriksson vs Marquese Chriss).
    """
    fa, la = split_name(a)
    fb, lb = split_name(b)
    first = jellyfish.jaro_winkler_similarity(fa, fb)
    last  = jellyfish.jaro_winkler_similarity(la, lb)
    return 0.35 * first + 0.65 * last

SCORE_THRESHOLD = 0.83
YEAR_WINDOW     = 1

# Build per-year buckets: year → list of (norm_name, original_name)
year_buckets = defaultdict(list)
for _, row in est_draft.iterrows():
    norm = normalize_name(row['Player'])
    for y in range(int(row['est_draft_year']) - YEAR_WINDOW,
                   int(row['est_draft_year']) + YEAR_WINDOW + 1):
        year_buckets[y].append((norm, row['Player']))

matches = []
for _, row in draft_df.iterrows():
    draft_name = row['PLAYER']
    draft_year = int(row['year'])
    norm_name  = normalize_name(draft_name)

    candidates = year_buckets.get(draft_year, [])
    if not candidates:
        matches.append({'PLAYER': draft_name, 'Player': None,
                        'match_score': 0.0, 'year_verified': False})
        continue

    scored = [(name_score(norm_name, c[0]), c[1]) for c in candidates]
    best_score, best_name = max(scored, key=lambda x: x[0])

    if best_score >= SCORE_THRESHOLD:
        matches.append({'PLAYER': draft_name, 'Player': best_name,
                        'match_score': round(best_score, 4), 'year_verified': True})
    else:
        matches.append({'PLAYER': draft_name, 'Player': None,
                        'match_score': round(best_score, 4), 'year_verified': False})

name_map = pd.DataFrame(matches)

unmatched = name_map['Player'].isna().sum()
print(f"Matched  : {len(name_map) - unmatched} / {len(name_map)}")
print(f"Unmatched: {unmatched}")
print()
print("Unmatched — sorted by closest miss (manual review candidates at top):")
name_map[name_map['Player'].isna()].sort_values('match_score', ascending=False).head(20)
```

```python
# Merge name map into draft_df, then join to totals
draft_mapped = draft_df.merge(name_map[['PLAYER', 'Player', 'match_score']], on='PLAYER', how='left')

# Join on the standardised Player name
df_joined = df.merge(
    draft_mapped.drop(columns=['PLAYER']),
    on='Player',
    how='left'
)

matched = df_joined['year'].notna().sum()
print(f"Rows with a draft match : {matched:,} / {len(df_joined):,}")
print(f"Shape of joined dataset : {df_joined.shape}")
df_joined.head(3)
```
